### Matplotlib Example

```python
# import and make data ready to plot
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4, 5]
y = [11, 22, 27, 44, 55]

# setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Example Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# save figure to a new location
fig.savefig("./data/images/example-plot.png")
```
