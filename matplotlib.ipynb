## Matplotlib Examples

```python
# import and make data ready to plot
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4, 5]
y = [11, 22, 27, 44, 55]

# setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Example Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# save figure to a new location
fig.savefig("./data/images/example-plot.png")
```

## Fig with Numpy Arrays

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# create data
x = np.linspace(0, 10, 100)
x[-10:]
```

```python
# plot data - line plots
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# plot data - scatter plots
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# plot data - another scatter plots
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# plot from dict - bar
butter_prices = {
    "Almond": 10,
    "Peanut": 4,
    "Cashew": 15
}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())

ax.set(
    title="Butter Prices",
    ylabel="Price"
);
```

```python
# horizontal bars
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()));
```

```python
# generate random data - histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### subplots - two options

```python
# option - 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 10))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), x[:10]/2);
ax3.bar(butter_prices.keys(), butter_prices.values())
ax4.hist(np.random.randn(1000));
```
